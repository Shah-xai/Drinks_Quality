```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import os
```

```python
os.chdir("../")
%pwd
```

```python
train_data = pd.read_csv('artifacts\\data_transformation\\train_transformed.csv')
train_data.head()
```

```python
model_1 = LogisticRegression(class_weight='balanced', max_iter=1000)
model_2 = SVC(class_weight='balanced', kernel='linear')
model_3 = RandomForestClassifier(class_weight='balanced', n_estimators=150)
model_4=SVC(class_weight='balanced', kernel='rbf')
X_train = train_data.drop('Potability', axis=1)
y_train = train_data['Potability']
```

```python
cv_scores_model_1 = cross_val_score(model_1, X_train, y_train, cv=5, scoring='roc_auc').mean()
cv_scores_model_2 = cross_val_score(model_2, X_train, y_train, cv=5, scoring='roc_auc').mean()
cv_scores_model_3 = cross_val_score(model_3, X_train, y_train, cv=5, scoring='roc_auc').mean()
cv_score_model_4 = cross_val_score(model_4, X_train, y_train, cv=5, scoring='roc_auc').mean()
print("Model 1 (Logistic Regression) CV Scores:", cv_scores_model_1)
print("Model 2 (SVC) CV Scores:", cv_scores_model_2)
print("Model 3 (Random Forest) CV Scores:", cv_scores_model_3)
print("Model 4 (SVC with RBF kernel) CV Scores:", cv_score_model_4)
```
